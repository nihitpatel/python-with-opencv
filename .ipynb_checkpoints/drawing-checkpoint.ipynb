```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = np.zeros((512,512,3),np.uint8)
plt.imshow(img)
plt.show()
```

```python
cv2.line(img,(0,0),(511,511),(0,0,255),5)
cv2.line(img,(0,511),(511,0),(0,0,255),5)
plt.imshow(img)
plt.show();
```

```python
cv2.rectangle(img,(200,200),(300,300),(255,0,0),5)
cv2.rectangle(img,(100,100),(400,140),(0,255,0),-1)
plt.imshow(img)
plt.show();
```

```python
cv2.circle(img,(256,256),80,(255,255,100),5)
i = 10
for i in range(0,256,30):
    cv2.circle(img,(256,256),i,(0,0,255),5)
plt.imshow(img)
plt.show()
```

```python
cv2.ellipse(img,(256,256),(150,70),50,0,360,(0,255,255),5)
plt.imshow(img)
plt.show()
```

```python
pts = np.array([[50,100],[150,200],[300,100],[350,200],[450,100]],np.int32)
#pts = pts.reshape((-1,1,2))
cv2.polylines(img,[pts],True,(0,0,0),5)
plt.imshow(img)
plt.show()
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'Nihit',(100,400),font,4,(0,0,0),5,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```
